# KNN - K Nearest Neighbours

While it can be used for Regression, it can perform very poorly, so we dont use it for Regression much.

## A **Classification** Algorithm

- Very simple
- Assigns a label to new data based on distance between old data and new data


## We take into account the distance between it and old data



### Significance of k

- Number of nearest neighbours to consider
- If one class is majority among the k, then the new point belongs to that class

### K = 3



### If ties occur

- Always choose odd k
- In case of tie, simple reduce k by 1 until tie is broken
- Choose nearest class point

### In scikit learn, we choose the nearest point



There could be a problem when we try to increase the value of k



The model could be aligned towards a class with more data points

### How to choose the best k then?

Choose a k that minimizes error
> Error = 1 - Accuracy

How?
- Elbow method
- Cros validate a grid search of multiple k values and choose a k that results in lowest error / highest accuracy

## Elbow method

- Train a model with multiple k values
- Measure the error

There are 2 elbow points here, one reduces error better than the other


We reach the elbow point, where increasing k does not significantly decrease error anymore. 

Its upto the human developer to choose the best k value depending upon how much it decreases error.

## Cross validation

Here we just perform cross validation until error is minimized

## General Working of KNN

- Choose K value
- Sort feature vectors by distance metric (From the label point)
- Choose class based on K nearest feature vectors

## Measuring distance metrics:

- Minkowski
- Euclidean
- Manhattan
- Chebyshev

**Note**:
> Features must be scaled. Very important for KNN
