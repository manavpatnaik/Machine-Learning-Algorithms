## Why Decision Trees are so well suited for Boosting?

### AdaBoost

- AdaBoost => Adaptive Boosting
- Uses by ensemble of weak learners
- Instead of developing trees in parallel, it develops trees in series
- Adapts by using previously created weak learners in order to adjust misclassified instances for the next created weak learner
- Hence the name: **AdaBoost**

### What is a weak learner?

- A model that is too simple to perform well on its own
- Example: A stump tree

- Learn slowly => A tree has to built before the next tree is built
