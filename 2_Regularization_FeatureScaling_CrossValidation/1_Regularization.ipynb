# Regularization

A way to
- Minimize model complexity
- Penalize loss functions
- **Reduce** model **overfitting** (by adding little more bias to reduce variance)

> In general, a way to reduce overfitting

> It requires a search for an optimal penalty hyperparameter

## 3 main types of Regularization

- L1 (LASSO Regression)
- L2 (Ridge Regression)
- Combining L1 and L2 (Elastic net)

## L1 Regularization

- Adds a penalty equal to absolute value of magnitute of coefficients
- Limits the size of coefficients

Helps identify which features are not usseful and which are not, it makes the very small coefficients equal to 0

Adds this penalty term to the loss/cost function, (p - number of coefficients, lambda - a hyperparameter that we can tune)


## L2 Regularization

Here
- Adds the square of magnitude of the coefficients
- Al coefficients are shrunk by same factor
- Does not necessarily eliminate coefficients like L1 (LASSO) Regularization

Adds this penalty term to the loss/cost function. (p, lambda - same as above)



#### Note:

> Basically both are trying to reduce the error by adding a penalty term to the loss/cost function

## Elastic Net

- Combines L1 and L2
- With an extra alpha parameter to decide the ratio betwen them



If alpha is
- 0: L2 Regularization
- 1: L1 regularization

## Cost of doing Regularization

- Introduces an additional hyperparameter that needs to be tuned
- Lambda - A multiplier to the penalty to decide the strength of penalty
