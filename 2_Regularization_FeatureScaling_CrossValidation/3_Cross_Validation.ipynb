# Cross validation

An advanced set of methods for splitting data into training and testing sets

Its a way to:
- Train on all data
- Evaluate on all data

Visual explanation:



The above splitting model is known as: **Train | Validation | Test split**



- Here we split the training data into **k** folds.
- Common value for **k = 10**
- We perform k iterations of training
- One of the folds wil be reserved for testing in each iteration
- While we train on each fold, we get a chance to tune our hyperparamaters



We take mean Error



> We get a much better sense of true performace

## Cost of doing cross validation

- High computation cost
- Computations are repeated **k times**
- We have to fairly tune the hyperparameters with every k fold

**Leave one out** cross validation: k = number of rows

Computationally very expensive
