# Feature Scaling

Some ML Models like KNN rely on distance metrics to perform well

**Why feature scaling?**
- Steepest descent algorithms do not possess **scale invariance**
- So **feature scaling** improves the **convergence** of those algorithms
- If features are on different scales, certain weights/coefficients may update faster than others

Feature scaling
- Affects direct interpretability of feature coefficients (makes it easier)
- Must always scale new unseen data before feeding to model

## Benefits

- Great increase in performance
- Absolutely necessary for some models
- No "real" downside to it

## 2 main ways

- **Standardization**: Rescales data to have mean=0 and SD=1
- **Normalization**: Rescales all data between 0 and 1

## Standardization
> Also called **z-score normalization**

(Not to be confused with normalization. Even if it is called z-score normalization, it is actually Standardization)

## Scikit Learn fit() and transform()

- fit(): Calculates necessary statistics: Xmin, Xmax, mean, std
- transform(): Converts the data as required


## Data leakage:

Calculating statistics from full data leads to some information of test set leaking into training process upon transform()

## Feature scaling process:

- Perform train test split
- Fit **only to training data**
- Transform training feature data
- Transform test feature data

#### Note:

> It is not advisable to scale the label

It can negatively impact Gradient Descent
