# Random Forest

- Another tree based method
- Has ability to greatly increase performance based on expanding ideas from Decision Trees

### Also known as _Ensemble Learner_

It relies on an ensemble of models (multiple decision trees)

## Problems with Decision Trees

Consider the dataset where the features are color coded.

### In a decision tree, the root node alone has a huge influence over the entire model



The operations that we can perform on this kind of tree
Adjust:
- Splitting Criterion (Information gain)
- Minimum Gini Impurity Decrease
- Setting depth limits
- Limits on number of leaf nodes

#### Limitations of a single decision tree:

- Single feature for root node
- Splitting criteria can lead to some features not being used
- Potential for overfitting of data

#### Solution:

- At every split: Randomly pick a subset of n number of features


- Build a decision tree based on those subset of features


- And again


### This way, we can ensure that all features are taken into consideration while building the tree. Set of multiple such trees are called Random Forests



### How to get a classification result out of it? Put it to vote



- P(class_0) = 2/3
- P(class_1) = 1/2

So Y => 0

### How to get a regression result out of it? Take mean



So Y => 11.3

It is proved mathematically that these random forests could allow trees to grow much deeper **without causing overfitting**

## Hyperparameters unique to Random Forests

(Most of the hyperparameters are same for Decision Trees and Random Forests)

Unique Hyperparameters
1. **Number of Estimators**
    - How many decision trees to use in the forest
    - No limit on number of trees
    - Adding more estimators does not result in overfitting
    - Default: 100 trees
    - Cross Validate a grid search of trees to choose a best number of trees
    - Plot Error vs Number of trees
    - As number of trees increase, trees become highly correlated
    - Different random selections are simply duplicating already created trees
    - Although we're not overfitting, we are not producing anything new
> "Random forests do no overfit. You can run as many trees as you want, it is fast." - LEO BREIMAN
2. **Number of features**
    - How many features to include in each subset
    - N - Total Number of features
    - Good starting point would be sqrt(N) or N/3
    
3. **Bootstrap samples** - Allow bootstrap sampling of each training subset
    - Bootstrapping: A term used to describe random sampling with replacement
    - Given a list of 5 letters: [A, B, C, D, E]
    - Bootstrapping: [ C, A, D, C, ... ] -> Random sampling with replacement
    - In Random Forest, we select a random subset of features at each split
    - Here bootstraping helps to further differentiate the trees
    - We could bootstrap a selection of rows for each split
    - 2 randomized training components in a tree
        - Subset of features used for splits
        - Bootstrapped rows of data
        
    - The main purpose is to reduce correlation between trees and  generalize the model more.
4. **Out-of-Bag error** - Calculate OOB during training?
    - Bagging: If we used Bootstrapping, some rows may not have been used for training
    - Rows 0,4,6 were not used during creation of decision trees
    
    - We could use those rows to come up with an error metric
    - It is known as OOB - Out of Bag Error
    
    - It doesnt affect training, just a performance metric
