## Main idea in SVM

**Use Hyperplanes** to create **separation** between classes

How to choose the best hyperplane?

### Solution: Use a separator that maximizes margins between classes



This is known as **Maximal Margin Classifier**

### What if classes are not perfectly separable?

- We cannot separate without some miscalculations
- We will have to have a BIAS-VARIANCE tradeoff




- Distance between threshold and observations is a **soft margin**
- Soft margins allow for misclassifications within the margins
