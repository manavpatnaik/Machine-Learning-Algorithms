# The chi squared $\chi$<sup>2</sup> distribution

- The $\chi$<sup>2</sup> distribution is a continuous prob. distribution of **sum of squares** of k independent **standard normal random variables**
- One of the most widely used distribution in inferential statistics and **hypothesis testing**

Z1 ... Zk are independent standard normal random variables


- Q is Chi-squared dist. 
- With **k degrees of freedom**. 
- K is the only parameter this distribution takes.


### Mean, variance & visualization of $\chi$<sup>2</sup> distribution

Consider 3 $\chi$<sup>2</sup> distributions

(Img src: jbstatistics)
- As the number of degrees of freedoms increase, the curve becomes more and more like a normal distribution.

Let **k = degrees of freedom**

> Mean = k

> Variance = 2k

> Mode = k - 2 (k >= 2)



## Why $\chi$<sup>2</sup> test?

- Primarily for **Hypothesis testing**
- Goodness of fit
- Test for independence
- Test of homogeneity

# Hypothesis testing

General Idea:
1. Make initial assumption
2. Collect data
3. Based on avalable data perform some tests to determine whether the initial hypothesis was valid or not

There are 2 hypotheses:
1. Null hypotheses
2. Alternate Hypotheses

## 1. The Null Hypothesis H<sub>0</sub>:

It is an assumption that there is no difference between the parameters under evaluation

For example: Consider a scenario where we are studying the gender population in a school

### The Null Hypothesis here would be:
- There are an equal number of students from all genders i.e No gender has more/less students than the other.

## 2. The Alternate Hypothesis H<sub>a</sub>:

- It states that there is significant relation between the variables / parameters.
- The significance level is generally taken as 0.05


> The $\chi$<sup>2</sup> test is one good way of performing hypothesis testing

## Formula:

In its raw form, the formula to calculate chi square is
> $\sigma$ ((observed - expected)<sup>2</sup> / expected)

# Calculation of Chi-Square


## 1. One way tables

Lets take an example of inheritance in **peas plants**:

Total sample size: 381

| Peas phenotype  | Observed Count  |
|---|---|
| Purple/Long  | 284  |
| Purple/Round  | 21  |
| Red/Long  | 21  |
| Red/Round  | 55  |

> H<sub>0</sub>: True ratio of phenotypes: 9:3:3:1

> H<sub>a</sub>: True ratio of phenotypes is not: 9:3:3:1

i.e H0 is: 
- P(Purple/Long) = 9/16
- P(Purple/Round) = 3/16
- ...

> Question: Are the observed counts significantly different from those expected under the NULL HYPOTHESIS?

### Expected counts under H<sub>0</sub>:

Total sample size: 381

| Peas phenotype  | Observed Count  |
|---|---|
| Purple/Long  | 381*(9/16) = 214.3  |
| Purple/Round  | 381*(3/16) = 71.4  |
| Red/Long  | 381*(3/16) = 71.4  |
| Red/Round  | 381*(1/16) = 23.8  |

### Observed and expected table


| Peas phenotype  | Observed Count  | Expected Count  |
|---|---|---|
| Purple/Long  | 284  | 214.3  |
| Purple/Round  | 21  | 71.4  |
| Red/Long  | 21  | 71.4  |
| Red/Round  | 55  | 23.8  |

### Calculating the chi_square value

Note:
- Degrees of freedon = # of categories - 1

1. If the observed values are going to be way different than expected, then the **chi-sq** is going to be high.
2. Large value of **chi-sq** is evidence against our Null Hypothesis

### For the above data: $\chi$<sup>2</sup> = 134.7

We need the p-value:

- DF: 4 - 1 = 3

p-value is the are to the right of our test statistic:


Its very close to 0.

### Inference: Since the p-value is very small, there is strong evidence against NULL HYPOTHESIS
