## Train | Test split

We train the model on the training data and evaluate it on the test data



**Procedure:**
1. Clean the data
2. Split the data into training and testing sets
3. Fit the scaler to the X training data
4. Scale the X test and X train data
5. Fit the model to the X train data
6. Test the model on the X test data
7. Calculate error, tune hyperparameter and then repeat steps 5 and 6

### Note: 

This method of evaluating the model and tuning the hyperparameters is not the best because we are adjusting the model based off error on the test set and re testing on the same data and are expecting improved performance.

(This is a fair technique. The only disadvantage is that the testing is not done on truly unseen data)

**Data Leaks** occur here.

## Data Leaks:

- 1st iteration:
    1. We test on test set and obtain error.
    2. We tune hyperparameters
- 2nd iteration:
    1. With tuned hyperparameters we test on the same test set
    
So some kind of **knowledge** is leaked from the test set and that knowledge is used to tune the hyperparameters. In that case we are bound to obtain a smaller error when testing on the test set again.

Due to this data leak, this kind of **training & testing** is not the best.

> We need data totally unseen/new to the model to test it on and know its performance

Enter **Validation Set**

## Train | Validation | Test split

If we want truly fair and final set of performance metrics, we need to have a test set that is set aside and is tested upon only at the end.

We split the data into 
1. Train
2. Validation
3. Test (aka Holdout set)



**Procedure:**
1. Fit the model on training set
2. Test on validation set
3. Adjust hyperparameters and repeat steps 1 and 2
4. Finally report the performance after a final testing on test set.

> This enables us to know the performance of the model on truly unseeen data


## k-Fold Cross Validation:

A variation of the train-validation-test split



**Procedure:**
1. Split the training data into k folds
2. Train the model on k-1 folds
3. Test the model on the kth fold
4. Tune hyperparameters, repeat steps 2 and 3 for different kth folds
5. Finally test in test set and report performance



For this approach
- As k increases, computation cost increases
- When we divide the training set into individual rows i.e k = len(train set), it is known as **leave one out cross validation**
- Leave one out CV: computationally most expensive

## CV in Scikit learn:

Many models perform k fold cross validation by default
- RidgeCV
- GridSearchCV
etc.

Scikit learn Process:
- Choose a k value
- Train on various combinations of folds
- Take avg error, adjust hyperparameters
- Obtain final metrics from the test set.

Functions that perform CV in scikit learn:
- cross_val_score()
- cross_validate()
