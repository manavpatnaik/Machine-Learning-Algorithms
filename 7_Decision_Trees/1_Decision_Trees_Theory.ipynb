## Gini Impurity

> A common information measurement for decision trees

- A mathematical measurement of how **pure** the dataset is
- For a set of classes **C** for a given dataset **Q**, Pc is prob of class **C**
> 
> 

- Gini impurity can be used to decide on data split values
- We want to minimize gini impurity at leaf nodes
- Minimized impurity at leaf nodes => We are separating classes effectively
