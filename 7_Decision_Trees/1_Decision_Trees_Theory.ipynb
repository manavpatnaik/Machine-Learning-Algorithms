## Gini Impurity

> A common information measurement for decision trees

- A mathematical measurement of how **pure** the dataset is

### Gini impurity for classification

- For a set of classes **C** for a given dataset **Q**, Pc is prob of class **C**

- Maximum impurity occurs when prob of bleonging to class is 0.5

### Consider a dataset with 4 points (2 red, 2 blue)




- Maximum possible gini impurity: 0.5
- If we replace a blue with red or red with blue, we make the dataset more pure w.r.t the representation of uniformity of classes. 
- Thereby decreasing impurity

### Lets swap one of the points (Data become more "pure")



### For a completely pure dataset

- Gini impurity = 0



### How is Gini impurity used in Decision Trees?

- Gini impurity can be used to decide on data split values
- We want to **minimize gini impurity at leaf nodes**
- Minimized impurity at leaf nodes => We are separating classes effectively

> The important ML task is minimizing gini impurities at the lead nodes
