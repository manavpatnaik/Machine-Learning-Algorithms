## OLS:

```y = mx + b``` => ```y = b0 + b1x```





## To calculate b0 and b1



## Limitations of Linear Regression: Anscombe's Quartet

### In practice

Lets have columns of data: x, y.

- x - feature
- y - label

````python
import numpy as np

# We train on this data
x = np.array([34, 35, 39, 42, 43, 47])
y = np.array([102, 109, 137, 148, 150, 158])

# This will be used for testing
test_set = (40, 134)

x_minus_xbar = x - np.mean(x)
y_minus_ybar = y - np.mean(y)

numerator = (x_minus_xbar * y_minus_ybar).sum()
denominator = ((x_minus_xbar)**2).sum()
````

````python
b1 = numerator/denominator
b1
````

````python
b0 = np.mean(y) - b1*np.mean(x)
b0
````

### For this dataset:

> y = -46 + 4.5*X

Now making our prediction

````python
def predict(x, b0, b1):
    return b0 + b1*x
````

````python
prediction = predict(test_set[0], b0, b1)
prediction
````

````python
print('Error in prediction: ', prediction - test_set[1])
````
